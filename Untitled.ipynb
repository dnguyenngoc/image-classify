```python
from helpers import image_utils

import matplotlib.pyplot as plt
plt.figure()

reader = easyocr.Reader(['vi', 'en'])
```

```python
giay_ra_vien = image_utils.load_datasets('../../datasets/image_classify/train/giay_ra_vien/')

import PIL
from PIL import ImageDraw
from PIL import Image
from matplotlib import cm
import numpy as np

# image = PIL.Image.open(giay_ra_vien[3])
image = image_utils.load(giay_ra_vien[3])
image = image_utils.pre_process(image)
plt.imshow(image, cmap='gray')
plt.show()
bounds = reader.readtext(image)


im = Image.fromarray(np.uint8(cm.gist_earth(image)*255))


# Draw bounding boxes
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for i in range(len(bounds)):
        bound = bounds[i]
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

new_image = draw_boxes(im, bounds)
plt.imshow(new_image, cmap='gray')
plt.show()
```
